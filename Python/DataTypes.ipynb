## Binary types

#### bytes
- A bytes object is an unchangeable sequence of integers. Once created, its content cannot be modified

```python
x = b'hello'
print(x)
print(type(x))
```

#### memoryview 
- A memoryview object allows you to access and manipulate the internal 
- data of an object (like bytes or bytearray) without making a copy

```python
y = memoryview(bytes(5))
print(y)
print(type(y))
```

#### Bytearray
- A bytearray is a mutable version of the bytes object. 
- we can modify individual bytes in the sequence without creating a new object

```python
z = bytearray(10)
print(z)
print(type(z))
```

## changing data types/ Type conversions/ type casting
#### Int to string

```python
n = 42
s = str(n)
print(n)
print(type(n))
print(s)
print(type(s)) # int to string
```

```python
n = [42,2]
s = f"{n}"
print(n)
print(type(n))
print(s)
print(type(s))  # list to string by using f'string
```

```python
n = [42,2]
s = repr(n)
print(n)
print(type(n))
print(s)
print(type(s))
```

```python
n = 'python'
s = list(n)
print(n)
print(type(n))
print(s)
print(type(s))
```

```python
n = 2
s = [n]
print(n)
print(type(n))
print(s)
print(type(s))
```

```python
a = [[0]*3]*3
print(a)
a[0][1] = 99 # All rows' first elements become 99
print(a) 

# fix 
a = [[5]*3 for _ in range(3)]
print(a)
```
